```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV
orbit = pd.read_csv("orbit.csv")

# Plot the orbit
plt.figure(figsize=(8, 8))

plt.plot(
    orbit["x"],
    orbit["y"],
    linewidth=2,
    label="Earth Orbit"
)

# Draw the Sun
plt.scatter(
    0,
    0,
    color="orange",
    s=120,
    label="Sun"
)

plt.xlabel("x (AU)")
plt.ylabel("y (AU)")
plt.title("Earth Orbit around the Sun")

# Equal scaling on both axes
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read simulation output
df = pd.read_csv("orbit.csv")

# Create figure
fig, axs = plt.subplots(2, 2, figsize=(14, 12))

# -------------------------------
# 1. Sun-Earth Orbits
# -------------------------------
# Earth
axs[0, 0].plot(
    df["earth_x"],
    df["earth_y"],
    linewidth=2,
    label="Earth"
)

# Sun
axs[0, 0].plot(
    df["sun_x"],
    df["sun_y"],
    linewidth=2,
    label="Sun"
)

# Earth start/end
axs[0, 0].scatter(
    df.iloc[0]["earth_x"],
    df.iloc[0]["earth_y"],
    color="green",
    s=50,
    label="Earth Start"
)

axs[0, 0].scatter(
    df.iloc[-1]["earth_x"],
    df.iloc[-1]["earth_y"],
    color="red",
    s=50,
    label="Earth End"
)

# Sun start/end
axs[0, 0].scatter(
    df.iloc[0]["sun_x"],
    df.iloc[0]["sun_y"],
    color="black",
    s=40,
    label="Sun Start"
)

axs[0, 0].scatter(
    df.iloc[-1]["sun_x"],
    df.iloc[-1]["sun_y"],
    color="magenta",
    s=40,
    label="Sun End"
)

# Center of Mass
axs[0, 0].scatter(
    df["center_of_mass_x"].iloc[0],
    df["center_of_mass_y"].iloc[0],
    marker="*",
    s=200,
    label="Center of Mass"
)

axs[0, 0].set_title("Sun-Earth Orbits")
axs[0, 0].set_xlabel("x (AU)")
axs[0, 0].set_ylabel("y (AU)")
axs[0, 0].axis("equal")
axs[0, 0].grid(True)
axs[0, 0].legend()

# -------------------------------
# 2. Energy
# -------------------------------
axs[0, 1].plot(df["kinetic_energy"], label="Kinetic")
axs[0, 1].plot(df["potential_energy"], label="Potential")
axs[0, 1].plot(df["total_energy"], linewidth=2, label="Total")

axs[0, 1].set_title("Energy")
axs[0, 1].set_xlabel("Time Step")
axs[0, 1].set_ylabel("Energy")
axs[0, 1].grid(True)
axs[0, 1].legend()

# -------------------------------
# 3. Angular Momentum
# -------------------------------
axs[1, 0].plot(df["angular_momentum_x"], label="Lx")
axs[1, 0].plot(df["angular_momentum_y"], label="Ly")
axs[1, 0].plot(df["angular_momentum_z"], linewidth=2, label="Lz")

axs[1, 0].set_title("Angular Momentum")
axs[1, 0].set_xlabel("Time Step")
axs[1, 0].set_ylabel("Angular Momentum")
axs[1, 0].grid(True)
axs[1, 0].legend()

# -------------------------------
# 4. Center of Mass
# -------------------------------
axs[1, 1].plot(df["center_of_mass_x"], label="COM x")
axs[1, 1].plot(df["center_of_mass_y"], label="COM y")
axs[1, 1].plot(df["center_of_mass_z"], label="COM z")

axs[1, 1].set_title("Center of Mass")
axs[1, 1].set_xlabel("Time Step")
axs[1, 1].set_ylabel("Position (AU)")
axs[1, 1].grid(True)
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
